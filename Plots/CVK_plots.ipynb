```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/Joined_table.csv", sep='\t')
```

```python
df = df[df['Область'] == 'Волинська область']
```

```python
df2 = df.groupby(['Партія', 'Статус'])['Партія'].count().unstack('Статус').dropna(axis=1, how='all')
df2 = df2[df2['обрано'].notna()]
df2 = df2.sort_values('обрано', ascending=True).reset_index()
df2 = df2.set_index('Партія')
ax = df2.plot(kind='barh', stacked=True, xlabel='')
ax.set_title('Волинська область', pad=20)
plt.style.use('seaborn-notebook')
plt.margins(0.6, 0.6)
plt.show()
```

```python
fig = ax.get_figure()
fig.savefig('fig.jpg', dpi=600, bbox_inches='tight')
```
